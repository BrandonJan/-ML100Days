### 1. [簡答題] 請問下列這三種方法有什麼不同？

```
print(a.sum()) 
print(np.sum(a))
print(sum(a)) 
```



````python
import numpy as np
a = np.arange(9).reshape(3,3)
print(a.sum()) 
print(np.sum(a))
print(sum(a)) 
print("a.sum() , np.sum(a) 輸出結果相同，皆是回傳所有加總的值")
print("sum(a)是以預設從 axis = 0 的方向做加總，得到一個一維的矩陣")
````

### 2. 請對一個 5x5 的隨機矩陣作正規化的操作。


````python
#正規化(normalization)是將特徵資料按比例縮放，讓資料落在某一特定的區間
import numpy as np
A = np.random.random((5, 5))
print(A)
B =  (A - A.min()) / (A.max() - A.min())
print("正規化後:")
print(B)

````

### 3. 請建立一個長度等於 10 的正整數向量，並且將其中的最大值改成 -1。


````python
import numpy as np
A = np.random.random(10)
A[A.argmax()] = -1
print(A)
````
