###  Q1: ÈÄèÈÅéÊï∏ÂÄºÊ≥ïË®àÁÆó  Age Âíå  Survived  ÊòØÂê¶ÊúâÁõ∏ÈóúÊÄß?


```python
#ÈÄ£Á∫åÂ∞çÈõ¢Êï£
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display

import pingouin as pg
import researchpy  
%matplotlib inline
df_train = pd.read_csv("Titanic_train.csv")
df_train['Survived_cate']=df_train['Survived']
df_train['Survived_cate']=df_train['Survived_cate'].astype(object)
#ÈÅãÁî® pg.anova Ë®àÁÆó‰∏âÁ®ÆËÆäÁï∞Êï∏
aov = pg.anova(dv='Age', between='Survived_cate', data=df_train, detailed=True)
print(aov)
#ËÆäÁï∞Êï∏ÊèõÁÆóÂæóÂà∞ Eta Squared (ùúº^ùüê)
etaSq = aov.SS[0] / (aov.SS[0] + aov.SS[1])
print(etaSq)

def judgment_etaSq(etaSq):
    if etaSq < .01:
        qual = 'Negligible'
    elif etaSq < .06:
        qual = 'Small'
    elif etaSq < .14:
        qual = 'Medium'
    else:
        qual = 'Large'
    return(qual)
print(judgment_etaSq(etaSq))
sns.set_style('whitegrid')
g = sns.catplot(x="Survived_cate", y="Age",data=df_train,kind='violin')
```

### Q2:ÈÄèÈÅéÊï∏ÂÄºÊ≥ïË®àÁÆó  Sex Âíå Survived  ÊòØÂê¶ÊúâÁõ∏ÈóúÊÄß?


```python
#Èõ¢Êï£Â∞çÈõ¢Êï£
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display

import pingouin as pg
import researchpy  
%matplotlib inline
df_train = pd.read_csv("Titanic_train.csv")
df_train['Survived_cate']=df_train['Survived']
df_train['Survived_cate']=df_train['Survived_cate'].astype(object)
#Áî®‰∫§ÂèâÂàóÈÄ£Ë°®(contingency table)Ôºå‰æÜÊï¥ÁêÜÂÖ©ÂÄãÈ°ûÂà•ÂûãÁöÑË≥áÊñô
contTable = pd.crosstab(df_train['Sex'], df_train['Survived_cate'])
print(contTable)
#Ë®àÁÆóË≥áÊñôËá™Áî±Â∫¶ df*
df = min(contTable.shape[0], contTable.shape[1]) - 1
print(df)
#ÈÅãÁî® researchpy Â•ó‰ª∂ÔºåË®àÁÆóÂá∫ Cramer‚Äôs V ‰øÇÊï∏
crosstab, res = researchpy.crosstab(df_train['Sex'], df_train['Survived_cate'], test='chi-square')
print(res)
print("Cramer's value is",res.loc[2,'results'])

#ÈÄôÈÇäÁî®Âç°ÊñπÊ™¢ÂÆöÁç®Á´ãÊÄßÔºåÊâÄ‰ª•Êé°Áî®ÁöÑ test ÂèÉÊï∏ÁÇ∫Âç°Êñπ "test =" argument.
# Êé°Áî®ÁöÑËÆäÊï∏Âú®ÈÄôÂÄãÊ®°ÁµÑ‰∏≠ÔºåÊúÉËá™Â∑±Ê†πÊìöË≥áÊñôÈõÜ‰æÜÂà§Êñ∑ÔºåCramer's Phi if it a 2x2 table, or Cramer's V is larger than 2x2.

## ÂØ´‰∏ÄÂÄãÂâØÁ®ãÂºèÂà§Êñ∑Áõ∏ÈóúÊÄßÁöÑÂº∑Â∫¶
def judgment_CramerV(df,V):
    if df == 1:
        if V < 0.10:
            qual = 'negligible'
        elif V < 0.30:
            qual = 'small'
        elif V < 0.50:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 2:
        if V < 0.07:
            qual = 'negligible'
        elif V < 0.21:
            qual = 'small'
        elif V < 0.35:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 3:
        if V < 0.06:
            qual = 'negligible'
        elif V < 0.17:
            qual = 'small'
        elif V < 0.29:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 4:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.15:
            qual = 'small'
        elif V < 0.25:
            qual = 'medium'
        else:
            qual = 'large'
    else:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.13:
            qual = 'small'
        elif V < 0.22:
            qual = 'medium'
        else:
            qual = 'large'
    return(qual)

print(judgment_CramerV(df,res.loc[2,'results']))
sns.set_style('whitegrid')
g= sns.countplot(x="Sex", hue="Survived_cate", data=df_train)
```

### Q3: ÈÄèÈÅéÊï∏ÂÄºÊ≥ïË®àÁÆó  Age Âíå  Fare ÊòØÂê¶ÊúâÁõ∏ÈóúÊÄß?  

```python
#ÈÄ£Á∫åÂ∞çÈÄ£Á∫å
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display

import pingouin as pg
import researchpy  
%matplotlib inline
df_train = pd.read_csv("Titanic_train.csv")
# Áî±Êñº pearsonr ÊúâÂÖ©ÂÄãÂõûÂÇ≥ÁµêÊûúÔºåÊàëÂÄëÂè™ÈúÄÂèñÁ¨¨‰∏ÄÂÄãÂõûÂÇ≥ÂÄºÁÇ∫Áõ∏Èóú‰øÇÊï∏
data=df_train[['Age','Fare']].dropna() #Ë®àÁÆóÁõ∏Èóú‰øÇÊï∏‰∏çËÉΩÊúâÈÅ∫Â§±ÂÄº
corr, _=stats.pearsonr(data['Age'], data['Fare'])
print(corr)
sns.set_style('whitegrid')
g = sns.regplot(x="Age", y="Fare", color="b",data=data)
```
